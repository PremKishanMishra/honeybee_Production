7) Find top 10 years with highest totalprod. [2Marks]

```python
# Group by year and sum the totalprod for each year
totalprod_by_year = data.groupby('year')['totalprod'].sum()

# Sort the values in descending order and get the top 10 years
top_10_years = totalprod_by_year.sort_values(ascending=False).head(10)

print("Top 10 years with the highest total production:")
print(top_10_years)
```

8) Find years with highest and lowest totalprod. [2Marks]

```python
# Find the year with the highest total production
year_highest_totalprod = totalprod_by_year.idxmax()
highest_totalprod = totalprod_by_year.max()

# Find the year with the lowest total production
year_lowest_totalprod = totalprod_by_year.idxmin()
lowest_totalprod = totalprod_by_year.min()

print("Year with the highest total production:")
print(f"Year: {year_highest_totalprod}, Total Production: {highest_totalprod}")

print("\nYear with the lowest total production:")
print(f"Year: {year_lowest_totalprod}, Total Production: {lowest_totalprod}")
```

9) Find top 10 states with highest totalprod. [2Marks]

```python
# Group by states and sum the totalprod for each year
totalprod_by_state = data.groupby('state')['totalprod'].sum()

# Sort the values in descending order and get the top 10 years
top_10_state = totalprod_by_state.sort_values(ascending=False).head(10)

print("Top 10 state with the highest total production:",top_10_state)


```

10) Show bottom 10 years with highest totalprod. [2Marks]

```python
# Sort the values in ascending order and get the bottom 10 years
bottom_10_years = totalprod_by_year.sort_values(ascending=True).head(10)

print("Bottom 10 years with the highest total production:")
print(bottom_10_years)
```

3) Find Skewness and Kurtosis of complete dataframe. [4 Marks]

4) Make a new copy of dataset and use for further analysis. [4 Marks]

```python
df_new=data.copy()
#using df_new as already copied the data in df at the starting to avoid any mistakes
```

```python


# Logarithmic transformation of skewed columns
skewed_columns = ['numcol', 'yieldpercol', 'priceperlb', 'stocks']
df_new_log_transformed = df_new[skewed_columns].apply(np.log1p)

# Visualize the transformed data
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, col in enumerate(skewed_columns):
    sns.histplot(df_new[col], kde=True, ax=axes[i // 2, i % 2], color='skyblue')
    axes[i // 2, i % 2].set_title(f'Distribution of {col}')

fig.suptitle('Original Skewed Data', fontsize=16)
plt.tight_layout()
plt.show()

# Visualize the transformed data
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, col in enumerate(skewed_columns):
    sns.histplot(df_new_log_transformed[col], kde=True, ax=axes[i // 2, i % 2], color='salmon')
    axes[i // 2, i % 2].set_title(f'Distribution of log({col} + 1)')

fig.suptitle('Transformed Data (Logarithmic Transformation)', fontsize=16)
plt.tight_layout()
plt.show()
```
